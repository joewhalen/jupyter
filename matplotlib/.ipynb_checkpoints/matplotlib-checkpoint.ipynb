## Let's generate some data

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
```

```python
def myAnnotate(axes,text,x0,y0,xf,yf,ypad = 0.05, capsize = 0.1,width=1, placement = 'above'):
    axes.annotate("", xy = (x0,y0), xytext = (xf,yf),
                 arrowprops=dict(arrowstyle='<->',
                                connectionstyle="arc3",
                                color='k',
                                shrinkA = 0, shrinkB = 0,lw=width))
    if placement == 'above:'
        axes.text((xf+x0)/2,(yf+y0)/2+ypad, text,  ha='center')
    elif placement == 'below':
        axes.text((xf+x0)/2,(yf+y0)/2+ypad, text,  ha='center')
    else:
        print('Ignoring invalid placement keyword')
        axes.text((xf+x0)/2,(yf+y0)/2+ypad, text,  ha='center')
    
    capkwargs = dict(color='k',lw=width,solid_capstyle='butt')
    ax.plot((x0,x0),(y0+capsize,y0-capsize), **capkwargs)
    ax.plot((xf,xf),(yf+capsize,yf-capsize),**capkwargs)
```

```python
ax = plt.subplot(111)
#ax.set_ylim(-1.5,1.5)
ax.plot(x,y)
myAnnotate(ax,'One Period',np.pi/2,1.15,5*np.pi/2,1.15)
ax.plot(x,y**2)
ax.patch
```
